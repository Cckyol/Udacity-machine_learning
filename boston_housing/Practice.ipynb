```python
from sklearn.linear_model import LinearRegression
from sklearn.learning_curve import learning_curve
import matplotlib.pyplot as plt 
from sklearn.metrics import explained_variance_score,make_scorer
from sklearn.cross_validation import KFold
import numpy as np
```

```python
size = 1000
cv_sets = KFold(size,shuffle=True)
```

```python
scorer = make_scorer(explained_variance_score)
```

```python
X = np.reshape(np.random.normal(scale=2,size=size),(-1,1))
y = np.array([[x[0]**2-2*x[0]+1] for x in X])
```

```python
reg = LinearRegression()
reg.fit(X,y)
print "Regressor score: {:.4f}".format(reg.score(X,y))
train_sizes,train_scores,test_scores = learning_curve(reg,X,y,cv=cv_sets,n_jobs=1,scoring="r2")    
```

```python
train_sizes,train_scores,test_scores = learning_curve(reg,X,y,cv=cv_sets,n_jobs=1,scoring="r2")
mean_train = np.mean(train_scores,axis = 1)
mean_test = np.mean(test_scores,axis =1)
plt.ylim(-0.1, 1.1)
plt.ylabel("Curve Score")
plt.xlabel("Training Points")
plt.legend(bbox_to_anchor=(1.1, 1.1))
plt.plot(train_scores,test_scores)
# plt.plot(train_scores)
print train_sizes
plt.plot(mean_train,mean_test)
plt.show()
```
